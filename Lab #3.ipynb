```python
# import Main Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Category Classification

### Problem 1

```python
X = np.array([[1,1],
            [2,1],
            [5,1],
            [6,1]])
X
```

```python
X.T
```

```python
X.T @ X
```

```python
X_inv = np.linalg.inv(X.T @ X)
X_inv
```

```python
X_F = X_inv @ X.T
X_F
```

```python
Y = np.array([-1,-1,1,1]).T
Y 
```

```python
W = X_F @ Y
W
```

```python
X_vals = X[:,0]
X_vals
```

```python
Y_per = W[0] * X_vals + W[1]
Y_per
```

```python
# plot
plt.plot(X_vals,Y_per)
plt.scatter(X_vals, Y, color='red')
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

### Problem 2

```python
X = np.array([[1,3,1],
            [2,4,1],
            [6,5,1],
            [7,4,1]])

Y = np.array([1,1,-1,-1]).T


X_inv = np.linalg.inv(X.T @ X)
X_F = X_inv @ X.T
W = X_F @ Y

X1_vals = X[:,0]
X2_vals = X[:,1]

Y_per = W[0] * X1_vals + W[1] * X2_vals + W[2]
```

```python
x_1 = np.linspace(2.5, 6, 100)
x_2 = - (W[0] * x_1 + W[2])/ W[1]
```

```python
# plot
plt.scatter(X2_vals, X1_vals, color='red')
plt.plot(x_1,x_2)
plt.plot()
plt.xlabel(r"$x_1$")
plt.ylabel(r"$x_2$")
plt.grid(True)
plt.show()
```
